```python
from transformers import pipeline
summarizer = pipeline("summarization", model = "facebook/bart-large-cnn")
text = "Electric cars transforming the industries"
print(summarizer(text, max_length=50, min_length=25))
```

```python
# ZERO SHORT PROMPTING

from transformers import pipeline
classifier = pipeline("zero-shot-classification" , model = "facebook/bart-large-mnli")

#input text, no examples given(zero short prompting)
sequence  = "I love learning Artificial Intelligence."

labels = ["sports", "technology", "education", "dancing"]

result = classifier(sequence, candidate_labels = labels)
print(result)
```

```python
# FEW SHORT PROMPTING

from transformers import pipeline

generator = pipeline("text-generation", model = "gpt2")

prompt = """
Translate the following English sentence into japanese

Examples:
1. Thank You  = arigatho
2. Good morning = ohayu

"""

result = generator(prompt, max_length = 3, do_sample=False)

print("Generated output", result[0]['generated_text'])
```
