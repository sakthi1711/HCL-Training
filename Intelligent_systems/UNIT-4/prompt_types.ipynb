```python
from openai import OpenAI
client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=os.environ["OPENAI_API_KEY"]
)
def ask_model(prompt, model="gpt-4o-mini"):
    response = client.chat.completions.create(
        model=model,  # you can change this to other OpenRouter models
        messages=[{"role": "user", "content": prompt}],
        temperature=0.7
    )
    return response.choices[0].message.content

```

```python
#zero shot

prompt = "Summarize this paragraph in one sentence:\nArtificial Intelligence is transforming industries by automating tasks, improving accuracy, and enabling data-driven decisions."
print(ask_model(prompt))
```

```python
#few shot

prompt = """You are a helpful assistant that summarizes text.

Example 1:
Input: The Earth revolves around the Sun once every 365 days.
Output: The Earth completes one orbit around the Sun each year.

Example 2:
Input: Photosynthesis allows plants to make food from sunlight.
Output: Plants use sunlight to produce food through photosynthesis.

Now summarize:
Artificial Intelligence is transforming industries by automating tasks, improving accuracy, and enabling data-driven decisions.
"""
print(ask_model(prompt))
```
